```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
sns.set_context("notebook", font_scale=1)
```

```python
symbol_images_dir = 'symbol_images/'
counter = {}
for tok in os.listdir(symbol_images_dir):
    imgs = os.listdir(symbol_images_dir + tok)
    counter[tok] = len(imgs)
print(f'{len(counter.keys())} classes, {sum(counter.values())} images')
```

```python
def name_mapping(ch):
#     if ch in '{}' or len(ch) >= 2:
#         if ch == 'sqrt':
#             ch = ch + r'{3}'
#         ch = '\\' + ch
#     ch = f'${ch}$'
    return ch
```

```python
data = sorted(counter.items(), key=lambda x: x[1], reverse=True)[:50]
x = [name_mapping(d[0]) for d in data]
y = [d[1] for d in data]
fig, ax = plt.subplots(figsize=(15,20))
bars = ax.barh(x, y)
ax.invert_yaxis()
ax.bar_label(bars)
plt.margins(y=0)
plt.savefig('figures/hw_symbols.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
